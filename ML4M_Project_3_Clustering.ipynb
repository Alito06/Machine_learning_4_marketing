# Data Understanding

## Dataset Descripiton


>- **ID** (*Numeric*) - Customer ID
- **Nationality**	 (*Categorical*) - Country of origin. Categories are represented in the ISO 3155–3:2013 format [1]|
- **Age**	(*Numeric*)	- Customer's age (in years) at the last day of the extraction period.
- **DaysSinceCreation**	(*Number*) -	Number of days since the customer record was created (number of days elapsed between the creation date and the last day of the extraction period)
- **NameHash**	(*Categorical*) -	Name of the customer's SHA2–256 hash string. A hash-string is the string resulting from a mathematical function that maps a string of arbitrary length to fixed-length [2]. Hash functions are used for different purposes. In this case, to allow customer's anonymization.
- **DocIDHash**	(*Categorical*)	- SHA2–256 hash-string of the identification document number the customer provided at check-in (passport number, national ID card number, or other)
- **AverageLeadTime**	(*Numeric*)	- The average number of days elapsed between the customer's booking date and arrival date. In other words, this variable is calculated by dividing the sum of the number of days elapsed between the moment each booking was made and its arrival date, by the total of bookings made by the customer.
- **LodgingRevenue**	(*Numeric*)	- Total amount spent on lodging expenses by the customer (in Euros). This value includes room, crib, and other related lodging expenses
- **OtherRevenue**	(*Numeric*)	- Total amount spent on other expenses by the customer (in Euros). This value includes food, beverage, spa, and other expenses
- **BookingsCanceled**	(*Numeric*)	- Number of bookings the customer made but subsequently canceled (the costumer informed the hotel he/she would not come to stay)
- **BookingsNoShowed**	(*Numeric*)	- Number of bookings the customer made but subsequently made a “no-show” (did not cancel, but did not check-in to stay at the hotel)
- **BookingsCheckedIn**	(*Numeric*)	- Number of bookings the customer made, and which end up with a staying
- **PersonsNights**	(*Numeric*)	- The total number of persons/nights that the costumer stayed at the hotel. This value is calculated by summing all customers checked-in bookings’ persons/nights. Person/nights of each booking is the result of the multiplication of the number of staying nights by the sum of adults and children
- **RoomNights**	(*Numeric*)	- Total of room/nights the customer stayed at the hotel (checked-in bookings). Room/nights are the multiplication of the number of rooms of each booking by the number of nights of the booking
- **DaysSinceLastStay**	(*Numeric*)	- The number of days elapsed between the last day of the extraction and the customer's last arrival date (of a checked-in booking). A value of −1 indicates the customer never stayed at the hotel
- **DaysSinceFirstStay**	(*Numeric*)	- The number of days elapsed between the last day of the extraction and the customer's first arrival date (of a checked-in booking). A value of −1 indicates the customer never stayed at the hotel
- **DistributionChannel**	(*Categorical*)	- Distribution channel usually used by the customer to make bookings at the hotel
- **MarketSegment**	(*Categorical*)	- Current market segment of the customer
SRHighFloor	Boolean	Indication if the customer usually asks for a room on a higher floor (0: No, 1: Yes)
-**SRHighFloor** (*Boolean*) - Indication if the customer usually asks for a room on a higher floor (0: No, 1: Yes)
- **SRLowFloor**	(*Boolean*)	- Indication if the customer usually asks for a room on a lower floor (0: No, 1: Yes)
- **SRAccessibleRoom**	(*Boolean*)	- Indication if the customer usually asks for an accessible room (0: No, 1: Yes)
- **SRMediumFloor**	(*Boolean*)	- Indication if the customer usually asks for a room on a middle floor (0: No, 1: Yes)
- **SRBathtub**	(*Boolean*)	- Indication if the customer usually asks for a room with a bathtub (0: No, 1: Yes)
- **SRShower**	(*Boolean*)	- Indication if the customer usually asks for a room with a shower (0: No, 1: Yes)
- **SRCrib**	(*Boolean*)	- Indication if the customer usually asks for a crib (0: No, 1: Yes)
- **SRKingSizeBed**	(*Boolean*)	- Indication if the customer usually asks for a room with a king-size bed (0: No, 1: Yes)
- **SRTwinBed**	(*Boolean*)	- Indication if the customer usually asks for a room with a twin bed (0: No, 1: Yes)
- **SRNearElevator**	(*Boolean*)	- Indication if the customer usually asks for a room near the elevator (0: No, 1: Yes)
- **SRAwayFromElevator**	(*Boolean*)	- Indication if the customer usually asks for a room away from the elevator (0: No, 1: Yes)
- **SRNoAlcoholInMiniBar**	(*Boolean*)	- Indication if the customer usually asks for a room with no alcohol in the mini-bar (0: No, 1: Yes)
- **SRQuietRoom**	(*Boolean*)	- Indication if the customer usually asks for a room away from the noise (0: No, 1: Yes)

## Outliers
Outliers refer to data points that deviate significantly from the remaining data. In our analysis, we have identified outliers that need attention as they can impact the results of clustering segmentation.

Firstly, in the variable "Age," we have observed both negative values like -11 and extreme values like 122. These values are unrealistic in the context of our data and need to be addressed to ensure accurate clustering.

Furthermore, when examining the summary statistics of the variables, we noticed that "AverageLeadTime" has a minimum value of -1, which is incorrect and requires correction. Additionally, negative values can be found in "DaysSinceLastStay" and "DaysSinceFirstStay," indicating instances where the guest has never stayed at the hotel.

Additionally, we have identified outliers in several other variables. These include "LodgingRevenue," "OtherRevenue," "BookingsCancelled," "BookingNoShowed," "BookingsCheckedIn," "PersonsNight," and "RoomsNight".

Addressing and handling these outliers is crucial to maintain the integrity and reliability of our clustering segmentation results.

```python
# Convert the negative values in the variable "Age" into positive ones
X['Age'] = X['Age'].abs()
X['AverageLeadTime'] = X['AverageLeadTime'].abs()
```

We decided to take the absolute values of these variables (i.e. convert negative values into positive ones), since we considered them to be data entry errors. For example, it's very likely that the hotel staff mistyped "11" and added a negative sign by mistake.

```python
# Fix remaining outliers in age

# Set the maximum plausible age limit
max_plausible_age = 100

# Identify the unrealistic values
unrealistic_values = X[X['Age'] > max_plausible_age]

# Replace the unrealistic values with the maximum plausible age
X.loc[X['Age'] > max_plausible_age, 'Age'] = max_plausible_age
```

## Feature Engineering

### Variable Creation

Since the variables with information about customer preferences contain very sparse data (most of them have a value of 0), this could make us fall into "the curse of dimensionality". As a result, our models will be very likely to find spurious relations between the variables and overfit. This is why we decided to combine all these variables into only one called 'ExpressedPreference'. In this way, we are simplifying the representation of customer preferences and reducing the number of dimensions in the dataset. This can help mitigate the risk of spurious relations and overfitting, making the models more interpretable and less prone to noise.

```python
# List of variables to combine
variables_to_combine = [
    'SRHighFloor',
    'SRLowFloor',
    'SRAccessibleRoom',
    'SRMediumFloor',
    'SRBathtub',
    'SRShower',
    'SRCrib',
    'SRKingSizeBed',
    'SRTwinBed',
    'SRNearElevator',
    'SRAwayFromElevator',
    'SRNoAlcoholInMiniBar',
    'SRQuietRoom'
]

# Apply the combining logic to create 'ExpressedPreference' column
X['ExpressedPreference'] = X[variables_to_combine].apply(lambda row: 1 if row.any() else 0, axis=1)

# Drop the columns used to create 'ExpressedPreference'
X = X.drop(variables_to_combine, axis=1)

# Print the updated DataFrame
print(X)
```

### One-Hot Encoding

Though adding new variables with many zeroes could create sparse data, we believe that information provided by the 'Nationality' variable is very valuable for our segmentation purposes. As it would be detrimental for our model to do one hot encoding for all the values in this variable, we decided to classify nationalities into 3 categories: 'Portugal', 'RestOfEurope' and 'OutOfEurope'. In this way, we can still retain some of the information provided by this variable without adding too much complexity to the model.

```python
# Perform one-hot encoding using get_dummies()
one_hot_encoded = pd.get_dummies(X['Nationality'], prefix='', prefix_sep='')

# Create new column 'Portugal' indicating if the nationality is 'PRT'
one_hot_encoded['Portugal'] = (X['Nationality'] == 'PRT').astype(int)

# Define the list of European countries
european_countries = ['FRA', 'PRT', 'DEU', 'GBR', 'ESP', 'ITA', 'BEL', 'NLD', 'CHE', 'IRL', 'AUT', 'SWE', 'NOR', 'POL', 'DNK', 'FIN', 'ROU', 'GRC', 'LUX', 'CZE', 'SVK', 'LVA', 'LTU', 'SVN', 'EST', 'HUN']

# Create new column 'RestOfEurope' indicating if the nationality is not 'PRT' and from Europe
one_hot_encoded['RestOfEurope'] = ((X['Nationality'] != 'PRT') & (X['Nationality'].isin(european_countries))).astype(int)

# Create new column 'OutOfEurope' indicating if the nationality is neither 'PRT' nor in Europe
one_hot_encoded['OutOfEurope'] = (~X['Nationality'].isin(['PRT'] + european_countries)).astype(int)

# Concatenate 'one_hot_encoded' with 'X'
X = pd.concat([X, one_hot_encoded[['Portugal', 'RestOfEurope', 'OutOfEurope']]], axis=1)

# Print the updated dataset
print(X)
```

These results represent the variance explained and cumulative variance explained by each principal component in a principal component analysis (PCA). PCA is a dimensionality reduction technique used to transform a set of correlated variables into a smaller set of uncorrelated variables called principal components.

- The "component" column indicates the index or number assigned to each principal component.

- Variance explained: This column shows the proportion of variance in the data explained by each principal component. It represents the amount of information captured by that component. In this case, the variance explained ranges from 0.000000 (component 10) to 0.301547(component 1). The higher the value, the more variance is explained by the corresponding component.

- Cumulative variance explained: This column shows the cumulative proportion of variance explained up to a specific principal component. It represents the cumulative amount of information captured by the combination of all previous components up to that point. In this case, the cumulative variance explained ranges from 0.301547 (component 1) to 1.000000 (component 11). The final value should always be 1.000000 (or 100%) since it represents the total variance in the original data.

To interpret the results, we can focus on the cumulative variance explained, and with this we can interpret the graph above. Looking into the variance curve plotted, and also having in mind that a value of above 85% of variance explained is considered good, we can argue that from the existence of 6 components our gains in including more component are residual. With 6 components the total variance explained is aproximatelly 0,94% which is a already quite high value of variance explained.

Knowing this informations is helpful in understanding the relative importance and contribution of each principal component in capturing the variability in the data.

The within-cluster sum of squares (WCSS) and inter-cluster distance are crucial metrics for evaluating the quality and characteristics of a clustering solution. A good clustering solution is characterized by a low WCSS, indicating compactness within clusters and data points that are closely grouped together.

Additionally, a desirable clustering solution exhibits a high inter-cluster distance, indicating distinct separation between clusters and well-differentiated clusters.

In our specific case, it is noteworthy that the intra-cluster distance, as represented by the WCSS, is slightly high. This can potentially be attributed to the observations within Cluster 1 and Cluster 2 being close to each other. For instance, there may be instances where two data points, each belonging to different clusters, are closer to each other than to their respective centroids.

Regarding the inter-cluster distance, the metric suggests that the model is moderately successful in distinguishing the different observations. While it may not exhibit a high degree of separation between clusters, it provides an average level of discrimination among the various observations.
